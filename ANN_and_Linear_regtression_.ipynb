```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as seaborn
```

```python
# Loading dataset from sklearn library
from sklearn.datasets import fetch_california_housing
```

```python
california = fetch_california_housing()
```

```python
california
```

```python
print(california.DESCR)
```

```python
# Create dataframe for my data
data=pd.DataFrame(california.data, columns=california.feature_names)
```

```python
data.head()
```

```python
data['Price']= california.target
data.head()
```

```python
# Data Preprocessing

#Step -2 Describe the data
data.describe()
```

```python
#To check null values
data.isna().sum()
```

```python
  # To check any duplicate values
data_dup=data.duplicated().sum() #checking the duplicate values
data_dup
```

```python
# Visualize outliers using box plots

plt.figure(figsize=(16, 8))
seaborn.boxplot(data, showfliers=True, flierprops=dict(markerfacecolor='red'))
plt.title("Boxplot of Features with Outliers Highlighted")
plt.xticks(rotation=45)
plt.show()
```

```python
# Checking Outliers
q3, q1 = data.quantile(0.75), data.quantile(0.25),
iqr = q3-q1
maxm, minm = q3 + 1.5*iqr, q1 - 1.5*iqr

print(((data < minm) | (data > (maxm))).sum())

data.head()
```

```python
# Removing outliers
data = data[~((data < minm) | (data > maxm)).any(axis=1)]
print(((data < minm) | (data > (maxm))).sum())
print("DataFrame shape after removing outliers:", data.shape)
```

```python
# Find correlation (for EDA)
data.corr()
```

```python
#creating a heatmap
seaborn.heatmap(data.corr(), annot=True)
```

```python

X=data.drop(["Price"], axis=1)
y=data['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train ,X_test, y_train, y_test = train_test_split(X,y, test_size=0.33 , random_state =42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
```

```python
X_train_sc = scaler.fit_transform(X_train)
X_test_sc = scaler.transform(X_test)
```

```python
# Model Training - Linear Regression

from sklearn.linear_model import LinearRegression
Reg = LinearRegression() # we can use any variable
```

```python
Reg.fit(X_train_sc, y_train)
```

```python

Reg.intercept_, Reg.coef_
```

```python
y_pred=Reg.predict(X_test_sc)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error,mean_squared_error
```

```python
# Calculation of Evaluation Metrics

print ("mse:", mean_squared_error(y_test, y_pred))
print ("mae: ", mean_absolute_error(y_test, y_pred))
print ("r2: ", r2_score(y_test, y_pred))
```

```python
# Scatter plot of actual vs predicted prices
plt.figure(figsize=(7.5, 7.5), tight_layout=True)
seaborn.scatterplot(x=y_test, y=y_pred, alpha=0.4, color='tab:blue')
plt.xlabel('Observed Price', fontsize=15)
plt.ylabel('Predicted Price', fontsize=15)
plt.title('Predicted vs Observed (Testing Data)', fontsize=18)

plt.xlim(0, max(y_test))

# 45-degree line starting from the origin (0, 0)
plt.plot([0, max(y_test)], [0, max(y_test)], color='tomato', ls='--', lw=2, label='45-deg line')

plt.legend(fontsize=15)
plt.show()
```

```python
# Importing necessary libraries for ANN model

from keras.models import Sequential
from tensorflow import keras
import tensorflow as tf
```

```python
# Build an ANN Model

model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(X_train_sc.shape[1],)),  # Input layer
    tf.keras.layers.Dense(units=64, activation='relu'),    # Hidden layer with ReLU activation
    tf.keras.layers.Dense(units=32, activation='relu'),    # Hidden layer with ReLU activation
    tf.keras.layers.Dense(units=1)                         # Output layer for regression
])
```

```python
# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Display the model architecture
model.summary()
```

```python
# Train the model
model.fit(X_train_sc, y_train, epochs=50, batch_size=32, verbose=1)

# Make predictions on the testing set
y_pred_ann = model.predict(X_test_sc).flatten()
```

```python
# Evaluate the model's performance using regression metrics

print ("mse_ann:", mean_squared_error(y_test, y_pred_ann))
print ("mae_ann: ", mean_absolute_error(y_test, y_pred_ann))
print ("r2_ann: ", r2_score(y_test, y_pred_ann))
```

```python
# Comparison of performance metrices of linear regression and ANN model

# Metrics for Linear Regression
mse = 0.3116
mae = 0.4258
r2 = 0.6417

# Metrics for ANN
mse_ann = 0.2026
mae_ann = 0.3255
r2_ann = 0.7669

print("Comparison of Performance Metrics:")
print("\nMetric                | Linear Regression | ANN")
print("--------------------- |-------------------|--------")
print(f"Mean Squared Error    | {mse:.4f}            | {mse_ann:.4f}")
print(f"Mean Absolute Error   | {mae:.4f}            | {mae_ann:.4f}")
print(f"R-squared (R2)        | {r2:.4f}            | {r2_ann:.4f}")
```

# Strengths and weaknesses of each model
Linear Regression

Strengths -
1. The interpretation of linear regression is simple and uncomplicated. Every aspect directly and clearly affects the anticipated result.
2. In particular on big datasets, linear regression is fast to train and computationally efficient.

Weakness -

1. The target variable and the features are assumed to have a linear relationship in linear regression. It might miss intricate, non-linear patterns in the data.
2. Complex associations are difficult for linear regression models to capture, particularly in datasets with elaborate patterns.

ANN Model -

Strengths-
1. ANNs are appropriate for tasks involving sophisticated patterns because they can capture complex, non-linear correlations in the data.
2. ANNs have the ability to automatically learn hierarchical feature representations, which may eliminate the requirement for manual  feature engineering.

Weakness

1. It can be computationally expensive to train ANNs, particularly for deep designs or huge datasets.
2. Since artificial neural networks (ANNs) are frequently regarded as "black box" models, it can be difficult to understand how they learn features and make decisions.

```python
# Adding complexity to ANN model to Analyse the model performance

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.regularizers import l2

# Increase the complexity of the ANN model with L2 regularization
model_complex_reg = Sequential([
    Dense(units=128, activation='relu', kernel_regularizer=l2(0.01), input_shape=(X_train_sc.shape[1],)),
    Dense(units=64, activation='relu', kernel_regularizer=l2(0.01)),
    Dense(units=32, activation='relu', kernel_regularizer=l2(0.01)),
    Dense(units=16, activation='relu', kernel_regularizer=l2(0.01)),
    Dense(units=1)
])

# Compile the model
model_complex_reg.compile(optimizer='adam', loss='mean_squared_error')

# Print the summary
model_complex_reg.summary()

```

```python
# Train the updated model
complex_reg = model_complex_reg.fit(X_train_sc, y_train, epochs=50, batch_size=32,  verbose=1)

# Make predictions on the testing set
y_pred_complex = model_complex_reg.predict(X_test_sc).flatten()

# Evaluate the model's performance using regression metrics
mse_complex_reg = mean_squared_error(y_test, y_pred_complex)
mae_complex_reg = mean_absolute_error(y_test, y_pred_complex)
r2_complex_reg = r2_score(y_test, y_pred_complex)

# Print the evaluation metrics
print("Mean Squared Error (MSE):", mse_complex_reg)
print("Mean Absolute Error (MAE):", mae_complex_reg)
print("R-squared (R2):", r2_complex_reg)
```

The performance of above model with L2 regularisation is slightly decreased because of several factors -
1. Overfitting could result from the increased complexity, particularly if the dataset is too small to accommodate the additional complexity. Overfitting is lessened by regularisation, although striking the correct balance is essential.
2. For the given dataset, the regularisation strength and other hyperparameters might not be ideal. For optimal performance, certain hyperparameters must be fine-tuned.


Consider If we add another complexity to our model
1. Increasing number of neurons - The network may capture more complex patterns by increasing the number of neurons in each layer.It may lead to better feature representation.
2.  Adjusting activation function - The model gains non-linearities from the use of several activation functions, which enable the model to learn more intricate correlations.
3. Increasing Training Epochs - The model can improve its parameters with additional training epochs.

# Conclusion

Key findings

1. R-squared and Mean Squared Error showed that the basic ANN performed better than Linear Regression.
2. An extra layer added to the ANN's complexity enhanced its performance even more.
3. Even though the complex ANN with L2 regularisation performs noticeably better than linear regression, it shows somewhat higher MSE and MAE than the basic ANN.


The Basic ANN outperformed Linear Regression, leveraging its ability to model non-linear relationships.


Challenges -

1. Model Complexity vs. Overfitting - It was difficult to strike a balance between the models' complexity and the possibility of overfitting. The simplest ANN first performed better, but as it became more complicated, overfitting became a trade-off.
2. Hyperparameter tuning - Finding optimal hyperparameters for both the ANN models was a time-consuming process. The choice of learning rate, batch size, and the number of epochs significantly impacted model performance
